```python
cache = []
for i in tqdm(prob_skus_ls):
  # fetch relevant sku #
  sample = df[df['sku']== i ]
  sample['New Category'] = sample['category_name_1']
  # find valid category other than backslash N #
  replace_val = sample[sample['category_name_1']!='\\N']['category_name_1'].values[0]
  # replace value where it make sense #
  sample.loc[sample['category_name_1']=="\\N",'New Category'] = replace_val

  # store back to df #
  cache.append(sample)
```

```python
len(cache)
```

```python
cache[0]
```

```python
prob_sku_df = pd.concat(cache, ignore_index=False)
```

```python
prob_sku_df.shape
```

```python
df.shape
```

```python
prob_sku_df['category_name_1'] = prob_sku_df['New Category']
prob_sku_df = prob_sku_df.drop(columns=['New Category'])
```

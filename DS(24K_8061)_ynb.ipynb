```python
#Activity 1
def calculator(a,b,operation):
  if operation == "addition" :
    return a+b
  elif operation == "subtraction":
    return a-b
  elif operation =="Multiplication" :
    return a*b
  elif operation == "division" :
     return a/b
  else:
    return "please try again!"
```

```python
calculator(10,20,"division")
```

```python
def lower(x):
  try:
    return str.lower()
  except:
    return x
```

```python
def check_gift(x):
  try:
    if 'BMW' in x:
      return 'car'
    else:
      return 'no car'
  except:
      return x
```

```python
def check(x):
    try:
        if  'wife' in x or 'daughter' in x or 'Wife' in x or 'Daughter' in x:
            return 'family'
        else:
            return 'others'
    except:
        return 'others'
```

```python
df['Name of Recipient'].apply(lambda x: check(x)).value_counts()
```

```python
df['Recipient Type'] = df['Name of Recipient'].apply(lambda x: check(x))
```

```python
recipient_type_counts = df['Recipient Type'].value_counts()
(recipient_type_counts)
```

```python
family_gifts = df[df['Recipient Type'] == 'family']
(family_gifts)
```

```python
family_gift_details = family_gifts[['Name', 'Detail of Gifts','Affiliation','Recipient Type']]
(family_gift_details)
```
